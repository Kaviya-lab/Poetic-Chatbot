```python
from transformers import pipeline
import time
```

```python
poetic_generator = pipeline("text-generation", model="gpt2")
```

```python
def dynamic_typing(text):
    """
    Prints the text one character at a time to simulate typing.
    """
    for char in text:
        print(char, end='', flush=True)
        time.sleep(0.03)  # Slightly faster typing effect
    print()  # New line at the end

# Step 5: Function to generate poetic response
def generate_poetry(user_input):
    """
    Takes user input and generates 2-6 lines of poetic text.
    """
    prompt = f"Turn this into a short poem: \"{user_input}\""

    result = poetic_generator(
        prompt,
        max_length=100,   # Adjust to control poem length
        num_return_sequences=1,
        do_sample=True,
        temperature=0.8   # Makes output more creative
    )

    poem = result[0]['generated_text']

    # Remove the original user input from output
    if poem.startswith(user_input):
        poem = poem[len(user_input):].strip()

    return poem

# Step 6: Interactive chatbot loop
print("Poetic Chatbot (type 'quit' to exit)")
while True:
    user_input = input("You: ")

    if user_input.lower() == "quit":
        print("Poetic Bot: Farewell, until next time! 🌙")
        break

    # Generate poetic response
    poem = generate_poetry(user_input)

    # Display poem with typing effect
    print("Poetic Bot: ", end="")
    dynamic_typing(poem)
```
